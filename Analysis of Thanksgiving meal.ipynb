```python
#Write a function to return the first integer value embedded in a string.
# (i) split a string on the space character
# (ii) replace a "+" with a blank character  
# (iii) return either (i) the integer value or (ii) None if null string entered
# Run a simple test on some string data

import pandas as pd
def str_to_int (input_str):
    if pd.isnull(input_str):
        return None
    else:
        string_list = input_str.split( )
        first_item = string_list[0]
        int_str = first_item.replace("+", "")
        return int(int_str)

#Test code for function above 
x = "68+ 987 098"
y = str_to_int(x)
print("Integer value = ", y)
x = "86 - 96"
y = str_to_int(x)
print("Integer value = ", y)
```

```python
#Write a function to return the first integer value embedded in a string.
# (i) split a string on the space character
# (ii) replace a "$" symbol with a blank character  
# (iii) return either (i) the integer value or (ii) None string is "Prefer" or Null

#Run the function on the Series data["How much total combined money did all members of \
# your HOUSEHOLD earn last year?"]

import pandas as pd

def str_to_int_2 (input_str):
    if pd.isnull(input_str): 
        return None
    else:
        string_list = input_str.split(" ")
        if string_list[0] == "Prefer":
            return None
        else:
            string_list = input_str.split(" ")
            int_str = string_list[0]
            int_str = int_str.replace("$", "")
            int_str = int_str.replace(",", "")
            return int(int_str)

data["int_income"] = data["How much total combined money did all members of \
your HOUSEHOLD earn last year?"].apply(lambda x:str_to_int_2(x))
data["int_income"].head(5)
data["int_income"].describe()
```

Comment/Anaylsis : Income level is take as the lower of the min/max levels and will skew the data accordingly. Average levels of income is relatively high and income have a large standard deviation.  

```python
#Select those with household incomes less than $150k 
income_lt_150k = data[data["int_income"] < 150000]
```

```python
#Summarise the total number of desserts
#Create two emply lists containing
# (i) the number of desserts consumed of each type and 
#  ... sum the number of non-null values in each column of "with_desserts" "with_deserts" 
# (ii) the name of each dessert  .. this is the text of first non null value in the same column

all_desserts_total, all_desserts_names = [], []
for i in range (0, len(with_desserts.iloc[0, :])):
    one_dessert_total = with_desserts.iloc[:,i].notnull().sum()
    all_desserts_total.append(one_dessert_total)
    all_desserts_names.append(with_desserts.iloc[:,i].dropna().iloc[0])

#Print out each list
print(all_desserts_total)
print(all_desserts_names)

```
